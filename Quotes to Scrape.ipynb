```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'http://quotes.toscrape.com/page/1/'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
quote_containers = soup.find_all('div', {'class': 'quote'})
```

```python
content = []

for quote in quote_containers:

    author_name = quote.find('small', {'class': 'author'}).text
    

    quote_text = quote.find('span', {'class': 'text'}).text
    

    tags = [tag.text for tag in quote.find_all('a', {'class': 'tag'})]

    content.append([author_name, quote_text, tags])
    print(content)
```

```python
data = pd.DataFrame(content,columns = ["author_name", "quote_text", "tags"])
```
